## 📊Introduction

Adolescent birth rate is an important health and social indicator that measures the number of births per 1,000 adolescent girls within a specific age group, usually between 15–19 years. High adolescent birth rates can affect the health, education, and economic opportunities of young women and may create social and economic challenges for communities and countries. Understanding adolescent birth patterns is essential for governments, healthcare organizations, and policymakers to improve healthcare systems and develop effective strategies.

In this project, the WHO Adolescent Birth Rate dataset is analyzed using Python-based data analytics techniques. The main objective of this analysis is to study adolescent birth trends across different countries and years, identify patterns and relationships in the data, and generate meaningful insights through Exploratory Data Analysis (EDA). The project involves data loading, cleaning, handling missing values, statistical analysis, and visualization methods such as bar charts, histograms, heatmaps, and line graphs.

The analysis helps in understanding variations in adolescent birth rates among countries and identifying trends over time. The findings from this study may support decision-making processes related to public health policies and future healthcare planning.

## ❓ Question 1: What is the statistical summary of the adolescent birth rate?

```python
df['rate_per_1k'].describe()
```

## 📌 Observation

The summary statistics show the average, minimum, maximum, and spread of adolescent birth rates across all countries and years.


## ❓ Question 2: How is the birth rate distributed?

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.histplot(df['rate_per_1k'], bins=30, kde=True)

plt.title('Distribution of  Birth Rate')
plt.xlabel('Birth Rate')
plt.ylabel('Frequency')

plt.show()
```

## 📌 Observation

The histogram helps identify whether the birth rate follows a normal distribution and reveals the presence of skewness or outliers.


## ❓ Question 3: Are there any outliers in adolescent birth rates?

```python
plt.figure(figsize=(8,5))

sns.boxplot(y=df['rate_per_1k'])

plt.title('Boxplot of Birth Rate')

plt.show()
```

## 📌 Observation

The boxplot highlights extreme birth rate values and helps identify countries with unusually high adolescent birth rates.

## ❓ Question 4: Which countries have the highest average adolescent birth rates?

```python
country_avg = df.groupby('GEO_NAME_SHORT')['rate_per_1k'].mean().sort_values(ascending=False)

top_10 = country_avg.head(10)
print(top_10)

import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
top_10.plot(kind='bar')

plt.title("Top 10 Countries by Average Adolescent Birth Rate")
plt.ylabel("Rate per 1k")
plt.xlabel("Country")

plt.show()
```

## 📌 Observation

A few countries have significantly higher adolescent birth rates than others, indicating regional disparities.


## ❓ Question 5: Which countries have the lowest average adolescent birth rates?

```python
country_avg = df.groupby('GEO_NAME_SHORT')['rate_per_1k'].mean().sort_values()

bottom10 = country_avg.head(10)

import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
bottom10.plot(kind='bar')

plt.title("Bottom 10 Countries by Average Birth Rate")
plt.ylabel("Rate per 1k")
plt.xlabel("Country")

plt.show()
```

## 📌 Observation

Several countries maintain consistently low adolescent birth rates, possibly due to better healthcare and education systems.

## ❓ Question 6: How has the average adolescent birth rate changed over time?

```python
yearly_birth_rate = df.groupby('year')['rate_per_1k'].mean()

yearly_birth_rate.head()

plt.figure(figsize=(12,6))

yearly_birth_rate.plot()

plt.title('Average Birth Rate Over Time')
plt.xlabel('Year')
plt.ylabel('Birth Rate')

plt.show()
```

## 📌 Observation
The trend line shows whether adolescent birth rates have increased or decreased over the years.

## ❓ Question 7: How do birth rates differ among age groups?

```python
df.groupby('age_group')['rate_per_1k'].mean()


plt.figure(figsize=(3,3))

sns.barplot(
    data=df,
    x='age_group',
    y='rate_per_1k'
)

plt.title('Average Birth Rate by Age Group')

plt.show()
```

## 📌 Observation

Different age groups show varying birth rate levels, helping identify the most affected age category.

## ❓ Question 8: What is the average birth rate for each age group and year?

#### Pivot Table

```python
pivot_table = pd.pivot_table(
    df,
    values='rate_per_1k',
    index='year',
    columns='age_group',
    aggfunc='mean'
)

pivot_table.head()
```

## 📌 Observation 

The pivot table provides a detailed comparison of birth rates across years and age groups.

## ❓ Question 9: What patterns can be observed using a heatmap?

```python
plt.figure(figsize=(12,6))

sns.heatmap(
    pivot_table,
    cmap='YlGnBu'
)

plt.title('Birth Rate Heatmap')

plt.show()
```

## 📌 Observation

The heatmap highlights years and age groups with high and low adolescent birth rates.

## ❓ Question 10: What is the country-wise statistical summary?

```python
country_summary = df.groupby('GEO_NAME_SHORT')['rate_per_1k'].agg(
    ['count', 'mean', 'median', 'min', 'max']
)

country_summary.head()
```

## 📌 Observation

Country-level statistics provide insights into the distribution of adolescent birth rates across nations.

## ❓ Q10: Subplots (Multiple Visuals Together)

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(2, 2, figsize=(14,10))

# 1️⃣ Histogram
sns.histplot(df['rate_per_1k'], kde=True, ax=axes[0,0])
axes[0,0].set_title("Distribution of Birth Rate")

# 2️⃣ Boxplot
sns.boxplot(y=df['rate_per_1k'], ax=axes[0,1])
axes[0,1].set_title("Boxplot of Birth Rate")

# 3️⃣ Top 5 countries
top5 = df.groupby('GEO_NAME_SHORT')['rate_per_1k'].mean().nlargest(5)

sns.barplot(
    x=top5.values,
    y=top5.index,
    ax=axes[1,0],
    color='steelblue'
)
axes[1,0].set_title("Top 5 Countries")

# 4️⃣ Bottom 5 countries
bottom5 = df.groupby('GEO_NAME_SHORT')['rate_per_1k'].mean().nsmallest(5)

sns.barplot(
    x=bottom5.values,
    y=bottom5.index,
    ax=axes[1,1],
    color='salmon'
)
axes[1,1].set_title("Bottom 5 Countries")

plt.tight_layout()
plt.show()
```
